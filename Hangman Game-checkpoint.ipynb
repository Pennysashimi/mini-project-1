```python

print("Welcome in the Hangman Game! \n")


while True: 
    chosen_dict = input("Choose your domain: \n For Words tap 1 \n For Movies tap 2 \n For JP tap 3 \n Your choice :  ") #fonction pour choisir le dictionnaire
    if chosen_dict == '1':
        selected = sample(list_words, 1)[0]
        print('\nYou choose Words.\n')
        break
    elif chosen_dict == '2':
        selected = sample(list_movies, 1)[0]
        print('\nYou choose Movies.\n')
        break
    elif chosen_dict == '3':
        selected = sample(list_JP, 1)[0]
        print('\nYou choose JP.\n')
        break
    else:
        print("\n Mmh, that'll be difficult. Please choose 1, 2 or 3.\n")
    
for l in selected: #fonction qui instaure une boucle pour ajouter autant de _ que de lettre dans le mot random au début
    if l in re.findall('[a-z]', selected): #limitation au caractère a-z
        display += "_ "# Remplacement des lettres par des "_ ""
    else:
        display += l #le reste s'affiche
        special_typeface.append(l)
        
        
while score_nb > 0: #fonction qui instaure une boucle tant que le score est supérieur à 0
    print("Guess:", display)
    letter = input('Tape your letter or word: ').lower() #variable qui définit la lettre choisie
    if letter == "quit":
        break
    if letter == selected:
        print(f"You win! Congratulations, you find the word {selected}.")
        break
    elif letter != selected and len(letter) != 1 :
        score_nb = score_nb - 1
        if score_nb == 1:
            print(hangman[score_nb], "Wrong word!.\nTry again.\nThis is your last chance.")
        elif score_nb == 0:
            break
        else:
            print(hangman[score_nb], f"Wrong word!\nTry again.\nYou still have {score_nb} attempts.")
    else:
        while  letter not in re.findall('[a-z]', letter): #Condition pour que la lettre choisie soit acceptable
            print('Try again.\nPlease choose just letter.')
            letter = input('Tape your letter or word: ').lower() #variable qui définit la lettre choisie
        if letter in selected: #boucle si la lettre proposée est dans la mot ou non 
            found_letters.append(letter)
            print("You\'re lucky!")
        else:
            score_nb = score_nb - 1
            if score_nb == 1:
                print(hangman[score_nb], "\nTry again.\nThis is your last chance.")
            elif score_nb == 0:
                break
            else:
                print(hangman[score_nb], f"\nTry again.\nYou still have {score_nb} attempts.")
        
    display = "" #Réinitialisation de l'affichage en _, avec remplacement de la lettre trouvée si dans selected 
    for x in selected:
        if x in special_typeface or x in found_letters:
            display += x
        else:
            display += "_ " 
            
    if "_" not in display: #Si toutes les lettres sont trouvées avant la fin 
        print(f"The word was {selected}. You win! Congratulations!")
        break
        
if score_nb == 0:
    print(hangman[score_nb], f"\nYou lost, sorry! The word was {selected}.")
print("END GAME")
```
